Cargo los datos en una lista de diccionarios y la transformo en un dataframe

```python
def cargar_lineas_json(ruta_archivo):
    data = []
    with open(ruta_archivo, 'r', encoding='utf-8') as f:
        for linea in f:
            try:
                # Convertir la línea en un diccionario y guardo en la lista
                linea_dict = json.loads(linea)
                data.append(linea_dict)
            except json.JSONDecodeError as e:
                print(f"Error al decodificar la línea: {linea}. Error: {e}")

    return pd.DataFrame(data)
```

Transformacion de los datos 


```python
# Convertimos la columna 'genres' a tipo lista, conservando NaN
df_steam_games['genres'] = df_steam_games['genres'].apply(lambda x: ast.literal_eval(x) if isinstance(x, str) else x)

# Expandimos las listas en filas
df_genres = df_steam_games.explode('genres')

# Obtener la lista de géneros únicos, incluyendo los valores nulos
lista_generos = df_genres['genres'].unique()

df_steam_games['tags'] = df_steam_games['tags'].apply(lambda x: ast.literal_eval(x) if isinstance(x, str) else x)
df_steam_games['genres'] = df_steam_games.apply(
    lambda row: [tag for tag in row['tags'] if tag in lista_generos] if isinstance(row['tags'], list) else row['genres'],
    axis=1
    )
```

```python
#Convertimos la columa release_date a formato fecha
df_steam_games['release_date']=pd.to_datetime(df_steam_games['release_date'], errors='coerce', exact=False)

#Solo nos interesa el año

# Extraer el año y crear una nueva columna 'release_year'
df_steam_games['year'] = df_steam_games['release_date'].dt.year.astype('Int64')

# Eliminar la columna 'release_date'
df_steam_games = df_steam_games.drop(columns=['release_date'])
```

```python
df_steam_games['year'] = df_steam_games['year'].fillna(df_steam_games['year'].median())
```

```python
# Convertir 'price' a tipo numérico, asignar NaN a 'Free To Play'
#df_steam_games['price'] = pd.to_numeric(df_steam_games['price'], errors='coerce')

# Reemplazar NaN por 0 en la columna 'price'
df_steam_games['price'].fillna(0, inplace=True)
```
